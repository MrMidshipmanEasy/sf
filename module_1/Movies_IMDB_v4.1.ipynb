# Предобработка

```python
answers = {} # создадим словарь для ответов

data['profit'] = data['revenue'] - data['budget']
data['date'] = data['release_date'].apply(lambda d: datetime.strptime(d, '%m/%d/%Y'))
data['date_int'] = data['date'].apply(lambda d: int(d.strftime("%Y%m%d")))
data['month'] = data['date'].apply(lambda d: d.month)


def unique_values(field):
    res = set()
    for gs in data[field].unique():
        for g in gs.split('|'):
            res.add(g)
    return list(res)

genres = unique_values('genres')
directors = unique_values('director')
artists = unique_values('cast')

```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = 'Daniel Radcliffe Rupert Grint'

h = {}



def build_key(name1, name2):
    l = [name1, name2]
    l.sort()
    return l[0]+' '+l[1]

def add(actor1, actor2):
    key = build_key(actor1, actor2)
    v = h.get(key,0)
    h[key] = v+1

for row in data['cast']:
    #print(row)
    actors  = row.split('|')
    l = len(actors)
    for i in range(l-1):
        for j in range(l-i-1):
            add(actors[i], actors[j+i+1])
            #print(actors[i], actors[j+i+1])
a = sorted(h.items(), key=lambda x: x[1], reverse=True)  

dict(a)
```
